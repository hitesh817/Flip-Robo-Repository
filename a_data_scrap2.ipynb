```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
drive=webdriver.Chrome(r'C:\Users\ADMIN00\Downloads\chromedriver_win32 (1)\chromedriver.exe')
drive.get("https://www.olx.in/items/q-hundai-cars?isSearchCall=true")
```

```python
title35=[]
price35=[]
location35=[]
brand35=[]
model35=[]
year35=[]
fuel35=[]
varient35=[]
km_driven35=[]
no_of_onwers35=[]
```

```python
for i in range(15):
    drive.find_element_by_xpath("//button[@data-aut-id='btnLoadMore']").click()
    time.sleep(8)
```

```python
urls7=[]
for i in drive.find_elements_by_xpath("//li[@data-aut-id='itemBox']/a"):
    urls7.append(i.get_attribute("href"))
urls7
```

```python
for q in urls7:
    drive.get(q)
    time.sleep(3)
    # fetch the title
    try:
        t35=drive.find_element_by_xpath("//h1[@class='_3rJ6e']")  
        title35.append(t35.text)
    except NoSuchElementException:
        title35.append('-') 
    # fetch the price
    try:
        p35=drive.find_element_by_xpath("//span[@class='_2xKfz']")  
        price35.append(p35.text)
    except NoSuchElementException:
        price35.append('-') 
    # fetch the location
    try:
        l35=drive.find_element_by_xpath("//div[@class='_2A3Wa']//span")  
        location35.append(l35.text)
    except NoSuchElementException:
        location35.append('-') 
    # fetch the brand
    try:
        b35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[1]/div[1]/span[2]")  
        brand35.append(b35.text)
    except NoSuchElementException:
        brand35.append('-') 
    # fetch the model
    try:
        m35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[2]/div[1]/span[2]")  
        model35.append(m35.text)
    except NoSuchElementException:
        model35.append('-') 
    # fetch the varient
    try:
        v35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[3]/div[1]/span[2]")  
        varient35.append(v35.text)
    except NoSuchElementException:
        varient35.append('-')
    # fetch the year
    try:
        y35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[4]/div[1]/span[2]")  
        year35.append(y35.text)
    except NoSuchElementException:
        year35.append('-')
    # fetch the fuel
    try:
        f35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[5]/div[1]/span[2]")  
        fuel35.append(f35.text)
    except NoSuchElementException:
        fuel35.append('-')
    # fetch the km driven
    try:
        d35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[7]/div[1]/span[2]")  
        km_driven35.append(d35.text)
    except NoSuchElementException:
        km_driven35.append('-')
    # fetch the no. of onwers
    try:
        ow35=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[8]/div[1]/span[2]")  
        no_of_onwers35.append(ow35.text)
    except NoSuchElementException:
        no_of_onwers35.append('-')

time.sleep(3)
```

```python
print(len(title35),len(price35),len(location35),len(brand35),len(model35),len(varient35),len(year35),len(fuel35),len(km_driven35),len(no_of_onwers35))
```

```python
car35= pd.DataFrame({'Title':title35,'Price':price35,'Location':location35,'Brand':brand35,'Model':model35,
                        'Year':year35,'Varient':varient35,'Fuel':fuel35,
                        'KM_Driven':km_driven35,'No_of_Onwers':no_of_onwers35})
car35
```

```python
drive.close()
```
