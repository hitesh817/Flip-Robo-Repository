```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
drive=webdriver.Chrome(r'C:\Users\ADMIN00\Downloads\chromedriver_win32 (1)\chromedriver.exe')
drive.get("https://www.olx.in/items/q-honda-city?isSearchCall=true")
```

```python
title30=[]
price30=[]
location30=[]
brand30=[]
model30=[]
year30=[]
fuel30=[]
varient30=[]
km_driven30=[]
no_of_onwers30=[]
```

```python
for i in range(10):
    drive.find_element_by_xpath("//button[@data-aut-id='btnLoadMore']").click()
    time.sleep(8)

```

```python
urls2=[]
for i in drive.find_elements_by_xpath("//li[@data-aut-id='itemBox']/a"):
    urls2.append(i.get_attribute("href"))
urls2
```

```python
for q in urls2:
    drive.get(q)
    time.sleep(3)
    # fetch the title
    try:
        t30=drive.find_element_by_xpath("//h1[@class='_3rJ6e']")  
        title30.append(t30.text)
    except NoSuchElementException:
        title30.append('-') 
    # fetch the price
    try:
        p30=drive.find_element_by_xpath("//span[@class='_2xKfz']")  
        price30.append(p30.text)
    except NoSuchElementException:
        price30.append('-') 
    # fetch the location
    try:
        l30=drive.find_element_by_xpath("//div[@class='_2A3Wa']//span")  
        location30.append(l30.text)
    except NoSuchElementException:
        location30.append('-') 
    # fetch the brand
    try:
        b30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[1]/div[1]/span[2]")  
        brand30.append(b30.text)
    except NoSuchElementException:
        brand30.append('-') 
    # fetch the model
    try:
        m30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[2]/div[1]/span[2]")  
        model30.append(m30.text)
    except NoSuchElementException:
        model30.append('-') 
    # fetch the varient
    try:
        v30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[3]/div[1]/span[2]")  
        varient30.append(v30.text)
    except NoSuchElementException:
        varient30.append('-')
    # fetch the year
    try:
        y30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[4]/div[1]/span[2]")  
        year30.append(y30.text)
    except NoSuchElementException:
        year30.append('-')
    # fetch the fuel
    try:
        f30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[5]/div[1]/span[2]")  
        fuel30.append(f30.text)
    except NoSuchElementException:
        fuel30.append('-')
    # fetch the km driven
    try:
        d30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[7]/div[1]/span[2]")  
        km_driven30.append(d30.text)
    except NoSuchElementException:
        km_driven30.append('-')
    # fetch the no. of onwers
    try:
        ow30=drive.find_element_by_xpath("//div[@class='_3JPEe']//div[8]/div[1]/span[2]")  
        no_of_onwers30.append(ow30.text)
    except NoSuchElementException:
        no_of_onwers30.append('-')

time.sleep(2)
```

```python
print(len(title30),len(price30),len(location30),len(brand30),len(model30),len(varient30),len(year30),len(fuel30),len(km_driven30),len(no_of_onwers30))
```

```python
car30= pd.DataFrame({'Title':title30,'Price':price30,'Location':location30,'Brand':brand30,'Model':model30,
                        'Year':year30,'Varient':varient30,'Fuel':fuel30,
                        'KM_Driven':km_driven30,'No_of_Onwers':no_of_onwers30})
car30 
```

```python
drive.close()
```
